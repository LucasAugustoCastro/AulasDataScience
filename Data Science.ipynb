```
def is_home_game(x): # valor de cada linha da coluna marchup
  if x.find('@') < 0: 
    return 1
  else: 
    return 0

# aplicar a função em cada valor da coluna matchup
df['home_game'] = df['matchup'].apply(is_home_game)
```

```
# usando apply com lambda
df['home_game_lambda'] = df['matchup'].apply(lambda x: 1 if (x.find('@') < 0) else 0)
```

```
# A partir de game_date, inferir anom mes, dia, dia da semana, e dia do ano
df['game_date'] = pd.to_datetime(df['game_date']) # sobrescrevendo a coluna original após conversao para tupo de data
df['ano'] = df['game_date'].dt.year
df['mes'] = df['game_date'].dt.month
df['dia'] = df['game_date'].dt.day
df['dia_da_semana'] = df['game_date'].dt.dayofweek
df['dia_do_ano'] = df['game_date'].dt.dayofyear 
df
```

```
# Criar a base
df = pd.DataFrame({
'x' : np.random.normal(10, 1.2, 20000),
'y' : np.random.normal(10, 1.2, 20000),
'group' : np.repeat('A', 20000)
})

tmp1 = pd.DataFrame({
'x' : np.random.normal(14.5, 1.2, 20000),
'y' : np.random.normal(14.5, 1.2, 20000),
'group' : np.repeat('B', 20000)
})

tmp2 = pd.DataFrame({
'x' : np.random.normal(9.5, 1.5, 20000),
'y' : np.random.normal(15.5, 1.5, 20000),
'group' : np.repeat('C', 20000)
})

# concatenating the data
df = df.append(tmp1).append(tmp2) 
```

```
df.head(5)
```

```
df.describe()
```

```
#analisando cada variavel de forma individuak
sns.distplot(df['x'])
plt.show()
```

```
#analisando cada variavel de forma individuak
sns.distplot(df['y'])
plt.show()
```

```
sns.scatterplot(df['x'], df['y'])
plt.show()
```

```
#mudar o tamanho do marcador
sns.scatterplot(df['x'], df['y'], s=3)
plt.show()
```

```
#transparencia
sns.scatterplot(df['x'], df['y'], alpha=0.01)
plt.show()
```

```
# densidade 

sns.kdeplot(df['x'], df['y'])
plt.show()
```

```
# densidade 

sns.kdeplot(df['x'], df['y'], shade=True)
plt.show()
```

```
# Join plot
sns.jointplot(df['x'], df['y'], kind='kde')
plt.show()
```

```
# Amostragem (sampling)
samp =df.sample(2000)
sns.scatterplot(samp['x'], samp['y'])
plt.show()
```

```
#enfase -> sar mais atençao a um dos grupos
df_a = df[df['group'] == 'A']
df_restante = df[df['group'] != 'A']

sns.scatterplot(df_restante['x'], df_restante['y'], color='grey')
sns.scatterplot(df_a['x'], df_a['y'], color='red')

plt.show()
```

```
# Facet grids -> facetamento
# analizar o comportamento de x e y, por grupo

g=sns.FacetGrid(df, col="group", hue='group')
g = g.map(plt.scatter, 'x','y')

plt.show();
```

```
# Criando nova base de dados
a = np.concatenate([
np.random.normal(2, 4, 1000),
np.random.normal(4, 4, 1000),
np.random.normal(1, 2, 500),
np.random.normal(10, 2, 500),
np.random.normal(8, 4, 1000),
np.random.normal(10, 4, 1000)
])

df = pd.DataFrame({
'x': np.repeat(range(1,6), 1000),
'y': a
}) 
```

```
df.head(5)
```

```
#scatterplot
sns.scatterplot(df['x'], df['y'])
plt.show();
```

```
# strpplot
sns.stripplot(df['x'], df['y'], jitter=0.3, size=2)
plt.show()
```

```
# combinando boxpltos com striplots
sns.boxplot(df['x'], df['y'])
sns.stripplot(df['x'], df['y'], jitter=0.3, size= 1, color='gray')

plt.show()
```

```
# combinando violinplot com striplots
sns.violinplot(df['x'], df['y'])
sns.stripplot(df['x'], df['y'], jitter=0.3, size= 1, color='gray')

plt.show()
```

```
# swarm plots 
sns.swarmplot(df['x'], df['y'])
plt.show()
```
